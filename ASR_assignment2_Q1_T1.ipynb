```python
### Download pre-trained model
import os
import tarfile
from coqui_stt_training.util.downloader import maybe_download

def download_pretrained_model():
    model_dir="english/"
    if not os.path.exists("english/coqui-yesno-checkpoints"):
        maybe_download("model.tar.gz", model_dir, "https://github.com/coqui-ai/STT-models/releases/download/english%2Fcoqui%2Fyesno-v0.0.1/coqui-yesno-checkpoints.tar.gz")
        print('\nNo extracted pre-trained model found. Extracting now...')
        tar = tarfile.open("english/model.tar.gz")
        tar.extractall("english/")
        tar.close()
    else:
        print('Found "english/coqui-yesno-checkpoints" - not extracting.')

# Download + extract pre-trained English model
download_pretrained_model()
```

```python
def extract_files():
  if os.path.exists("/content/marathi/data/wavs/test"):
    print("Data already extracted!")
  else:
    print("Extracted data not found, extracting")
    maybe_download("data.tgz","marathi/", "https://www.cse.iitb.ac.in/~pjyothi/cs753/data.tgz")
    tar = tarfile.open("marathi/data.tgz")
    tar.extractall("marathi/")

extract_files()
```

```python
from pathlib import Path
import glob

def get_info(mypath):
  files = [Path(f).stem for f in glob.glob(f'{mypath}/*')]
  sizes = [os.path.getsize(f) for f in glob.glob(f'{mypath}/*')]
  info = dict(zip(files, sizes))
  return info
```

```python
import pandas as pd 

def create_csv(mypath, ftype):
  info = get_info(mypath)
  tsv_file = 'marathi/data/marathi.tsv'
  csv_table = pd.read_table(tsv_file,sep='\t',names=['wav_filename', 'transcript'])
  csv_table = csv_table[csv_table['wav_filename'].isin(list(info.keys()))].reset_index(drop=True)

  size = pd.Series([])
  for i in range(len(info)):
    size[i] = info[csv_table['wav_filename'][i]]

  csv_table.insert(1, 'wav_filesize', size)
  csv_table['wav_filename'] = f'{ftype}/' + csv_table['wav_filename'].astype(str) + '.wav'
  csv_table.to_csv(f'{mypath}_marathi.csv',index=False)
```

```python
create_csv('marathi/data/wavs/test', 'test')
create_csv('marathi/data/wavs/train', 'train')
```
