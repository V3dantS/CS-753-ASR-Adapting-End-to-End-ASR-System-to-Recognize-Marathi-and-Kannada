```python
with open("data/train.tsv","r") as fp:
    all_text = fp.readlines()
```

```python
languages = set([x.split('\t')[0].split('_')[0] for x in all_text])
```

```python
text_lang = {}
for lang in languages:
    lang_text = list(filter(lambda x:lang in x, all_text))
    text_lang[lang]=lang_text
```

```python
kannada_text = [x.strip().split('\t')[1] for x in text_lang['knf']]
kannada_alphabets = list(chain.from_iterable(kannada_text))
```

```python
telugu_text = [x.strip().split('\t')[1] for x in text_lang['tef']]
telugu_alphabets = list(chain.from_iterable(telugu_text))
```

```python
tamil_text = [x.strip().split('\t')[1] for x in text_lang['taf']]
tamil_alphabets = list(chain.from_iterable(tamil_text))
```

```python
malay_text = [x.strip().split('\t')[1] for x in text_lang['mlf']]
malay_alphabets = list(chain.from_iterable(malay_text))
```

```python
with open('data/telugu_alphabet.txt','w') as fp:
    for alphabet in set(telugu_alphabets):
        fp.write(alphabet)
        fp.write('\n')
with open('data/tamil_alphabet.txt','w') as fp:
    for alphabet in set(tamil_alphabets):
        fp.write(alphabet)
        fp.write('\n')
with open('data/malay_alphabet.txt','w') as fp:
    for alphabet in set(malay_alphabets):
        fp.write(alphabet)
        fp.write('\n')
with open('data/kannada_alphabet.txt','w') as fp:
    for alphabet in set(kannada_alphabets):
        fp.write(alphabet)
        fp.write('\n')
```

```python
from pathlib import Path
import glob

def get_info(mypath):
  files = [Path(f).stem for f in glob.glob(f'{mypath}/*')]
  sizes = [os.path.getsize(f) for f in glob.glob(f'{mypath}/*')]
  info = dict(zip(files, sizes))
  return info
```

```python
import pandas as pd 

def create_csv(mypath, ftype):
  info = get_info(mypath)
  tsv_file = '/content/data/train.tsv'
  csv_table = pd.read_table(tsv_file,sep='\t',names=['wav_filename', 'transcript'])
  csv_table = csv_table[csv_table['wav_filename'].isin(list(info.keys()))].reset_index(drop=True)

  size = pd.Series([])
  for i in range(len(info)):
    size[i] = info[csv_table['wav_filename'][i]]

  csv_table.insert(1, 'wav_filesize', size)
  csv_table['wav_filename'] = f'{ftype}/' + csv_table['wav_filename'].astype(str) + '.wav'
  csv_table.to_csv(f'{mypath}_marathi.csv',index=False)
```
